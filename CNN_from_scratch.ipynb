```python
data = pd.read_csv('/content/train.csv')
data.head()
```

```python
data = np.array(data)
m,n = data.shape
np.random.shuffle(data)
```

```python
data_train = data[0:7000].T
x_train = data_train[1:n]
y_train = data_train[0]
```

```python
data_dev = data[7000:].T
x_dev = data_dev[1:n]
y_dev = data_dev[0]
```

```python
kernel = np.array([
    [1,0,-1],
    [1,0,-1],
    [1,0,-1],
])
```

```python
x_train.shape
```

```python
x_train = x_train.reshape(7000, 28, 28)
```

```python
row_kernel, col_kernel = kernel.shape
output = np.ones((7000, 28-row_kernel+1, 28-col_kernel+1))
```

```python
for i in range(len(x_train)):
    for j in range(28-row_kernel+1):
        for k in range(28-col_kernel+1):
            output[i][j][k] = np.sum(x_train[i][j:j+row_kernel, k:k+col_kernel]*kernel)
```
